```python
#data preprocessing
#select  an appropriate sample[male, 200 features only]
male_df = data[data['Gender'] == 'Male'][:200]
male_df
```

convert height(inches to cm) and weight(pounds to kg)
1 pound = 0.453592kg
1 inch = 2.54cm

```python
male_df['Height'] = male_df['Height'].apply(lambda x: x*2.54)
male_df['Weight'] = male_df['Weight'].apply(lambda x: x*0.453592)
male_df
```

```python
#define the gradient
def line_of_best_fit(xs, ys):
    slope = ((mean(xs)*mean(ys))-mean(xs*ys))/((mean(xs)*mean(xs))-mean(xs*xs))
    y_intercept = mean(ys) - slope*mean(xs)
    return slope, y_intercept
```

```python
#convert the column data into a list for ease fo interpretation by python
height_list = male_df['Height'].tolist()
weight_list = male_df['Weight'].tolist()
```

```python
#define which of the variables height and weight will be x and y
#1) height_list == xs
#2) weight_list == ys
#3) these values are numerical so we have to import the numpy library
```

```python
import numpy as np
```

```python
xs =np.array(height_list, dtype = np.float64)
ys =np.array(weight_list, dtype = np.float64)
```

```python
#plot the graphics
#1) we will use the matplotlib library to display the curve
#2) we will use the mean from statistics library to display the mean values int the graph
```

```python
import matplotlib.pyplot as plt
from statistics import mean
from matplotlib import style
```

```python
#plot the graph
style.use('seaborn')
plt.scatter(xs,ys, label = 'Data Points', alpha = 0.6, color='green', s = 75)
plt.title("Scatter plot of height(cm) against weight(kg)")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
```

Model building

```python
# the model will be a llinear regressional model
# STEPS
# 1) calculate the slope
# 2) get the linear regression calculated from the slope and the y_intercept
# 3) make a prediction
```

```python
# 1) calculate the slope
slope, y_intercept = line_of_best_fit(xs, ys)
```

```python
# 2) get the linear regression calculated from the slope and the y_intercept
regression_line = [(slope*x) + y_intercept for x in xs]
```

```python
# 3) make a prediction
#What is the weight when the height is 186.25 cm?
man_height = 186.25
man_weight = (slope*man_height) + y_intercept
```
